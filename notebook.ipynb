```python
Missing = ['PM25', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
changping_data_df[Missing] = changping_data_df[Missing].fillna(changping_data_df[Missing].interpolate())
```

##### Outlier

```python
def Maximum(df, column):
    Q1 = (df[column].quantile(0.25))
    Q3 = (df[column].quantile(0.75))
    IQR = Q3 - Q1

    maximum = Q3 + (1.5*IQR)

    maximum = df[column] > maximum

    return maximum

def Minimum(df, column):
    Q1 = (df[column].quantile(0.25))
    Q3 = (df[column].quantile(0.75))
    IQR = Q3 - Q1

    minimum = Q1 - (1.5*IQR)

    minimum = df[column] < minimum

    return minimum
```

```python
lower_PM25 = Minimum(changping_data_df, 'PM25')
more_PM25 = Maximum(changping_data_df, 'PM25')

changping_data_df = changping_data_df.assign(
    PM25 = lambda x: x['PM25'].mask(
        (x['PM25'] < lower_PM25) | (x['PM25'] > more_PM25),
        x['PM25'].rolling(window=10, min_periods=1).mean()
    )
)
```

```python
lower_PM10 = Minimum(changping_data_df, 'PM10')
more_PM10 = Maximum(changping_data_df, 'PM10')

changping_data_df = changping_data_df.assign(
    PM10 = lambda x: x['PM10'].mask(
        (x['PM10'] < lower_PM10) | (x['PM10'] > more_PM10),
        x['PM10'].rolling(window=10, min_periods=1).mean()))
```

```python
lower_SO2 = Minimum(changping_data_df, 'SO2')
more_SO2 = Maximum(changping_data_df, 'SO2')

changping_data_df = changping_data_df.assign(
    SO2 = lambda x: x['SO2'].mask(
        (x['SO2'] < lower_SO2) | (x['SO2'] > more_SO2),
        x['SO2'].rolling(window=10, min_periods=1).mean()
    )
)
```

```python
lower_CO = Minimum(changping_data_df, 'CO')
more_CO = Maximum(changping_data_df, 'CO')

changping_data_df = changping_data_df.assign(
    CO = lambda x: x['CO'].mask(
        (x['CO'] < lower_CO) | (x['CO'] > more_CO),
        x['CO'].rolling(window=10, min_periods=1).mean()
    )
)
```

### Merge data

```python
merged_df1 = pd.merge(changping_data_df, dongsi_data_df, on='datetime', suffixes=('_changping', '_dongsi'))
merged_df2 = pd.merge(guanyuan_data_df, gucheng_data_df, on='datetime', suffixes=('_guanyuan', '_gucheng'))
all_country_df = pd.merge(merged_df1, merged_df2, on='datetime')

average_df = pd.DataFrame()
average_df['No'] = all_country_df['No_changping']
average_df['datetime'] = all_country_df['datetime']
# average_df['year'] = all_country_df['year_changping']
# average_df['month'] = all_country_df['month_changping']
# average_df['day'] = all_country_df['day_changping']
# average_df['hour'] = all_country_df['hour_changping']
average_df['PM25'] = round((all_country_df[['PM25_changping', 'PM25_dongsi', 'PM25_guanyuan', 'PM25_gucheng']].mean(axis=1)), 2)
average_df['PM10'] = round((all_country_df[['PM10_changping', 'PM10_dongsi', 'PM10_guanyuan', 'PM10_gucheng']].mean(axis=1)), 2)
average_df['SO2'] = round((all_country_df[['SO2_changping', 'SO2_dongsi', 'SO2_guanyuan', 'SO2_gucheng']].mean(axis=1)), 2)
average_df['NO2'] = round((all_country_df[['NO2_changping', 'NO2_dongsi', 'NO2_guanyuan', 'NO2_gucheng']].mean(axis=1)), 2)
average_df['CO'] = round((all_country_df[['CO_changping', 'CO_dongsi', 'CO_guanyuan', 'CO_gucheng']].mean(axis=1)), 2)
average_df['O3'] = round((all_country_df[['O3_changping', 'O3_dongsi', 'O3_guanyuan', 'O3_gucheng']].mean(axis=1)), 2)
average_df['TEMP'] = round((all_country_df[['TEMP_changping', 'TEMP_dongsi', 'TEMP_guanyuan', 'TEMP_gucheng']].mean(axis=1)), 2)
average_df['PRES'] = round((all_country_df[['PRES_changping', 'PRES_dongsi', 'PRES_guanyuan', 'PRES_gucheng']].mean(axis=1)), 2)
average_df['DEWP'] = round((all_country_df[['DEWP_changping', 'DEWP_dongsi', 'DEWP_guanyuan', 'DEWP_gucheng']].mean(axis=1)), 2)
average_df['RAIN'] = round((all_country_df[['RAIN_changping', 'RAIN_dongsi', 'RAIN_guanyuan', 'RAIN_gucheng']].mean(axis=1)), 2)
average_df['WSPM'] = round((all_country_df[['WSPM_changping', 'WSPM_dongsi', 'WSPM_guanyuan', 'WSPM_gucheng']].mean(axis=1)), 2)

average_df.sample(10)
```

### Pertanyaan 1: Bagaimana grafik rata-rata dari masing-masing parameter pencemar udara di keempat kota tersebut?

```python
monthly_date_df = average_df.resample(rule='M', on='datetime').agg({
    'No': 'nunique',
    'PM25': 'mean',
    'PM10': 'mean',
    'SO2': 'mean',
    'NO2': 'mean',
    'CO': 'mean',
    'O3': 'mean',
    'TEMP': 'mean',
    'PRES': 'mean',
    'DEWP': 'mean',
    'RAIN': 'mean',
    'WSPM': 'mean'
})
monthly_date_df.index = monthly_date_df.index.strftime('%B %Y')
monthly_date_df = monthly_date_df.reset_index()
monthly_date_df.rename(columns={
    'datetime': 'Bulan',
    'No': 'Jumlah'
}, inplace=True)
monthly_date_df.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(
    monthly_date_df['Bulan'],
    monthly_date_df['TEMP'],
    marker='o',
    color='#006A67',  # Warna garis untuk PM10
    linewidth=2,
    label='Temperature'
)
plt.title('Rata-rata temperature per bulan (2013 - 2017)', loc='center', fontsize=20)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 7.5))
plt.plot(
    monthly_date_df['Bulan'],
    monthly_date_df['PM25'],
    marker='o',
    color='#000B58',  # Warna garis untuk PM2.5
    linewidth=2,
    label='Mean PM2.5'
)

plt.plot(
    monthly_date_df['Bulan'],
    monthly_date_df['PM10'],
    marker='o',
    color='#006A67',  # Warna garis untuk PM10
    linewidth=2,
    label='Mean PM10'
)
plt.title('Mean of PM2.5 and PM10 per Month (2013 - 2017)', loc='center', fontsize=20)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.show()
```

```python
sns.set(style='darkgrid')

plt.figure(figsize=(12,6))

sns.lineplot(data=monthly_date_df, x='Bulan', y='PM25', label='PM2.5', color='blue')
sns.lineplot(data=monthly_date_df, x='Bulan', y='PM10', label='PM10', color='green')
sns.lineplot(data=monthly_date_df, x='Bulan', y='SO2', label='SO2', color='red')
sns.lineplot(data=monthly_date_df, x='Bulan', y='NO2', label='NO2', color='purple')
sns.lineplot(data=monthly_date_df, x='Bulan', y='O3', label='O3', color='orange')
sns.lineplot(data=monthly_date_df, x='Bulan', y='TEMP', label='TEMP', color='brown')

plt.xlabel('Bulan')
plt.ylabel('Konsentrasi (µg/m)³')
plt.title('Rata-rata Pencemar udara setiap bulan')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
