##### Outlier

```python
def Maximum(df, column):
    Q1 = (df[column].quantile(0.25))
    Q3 = (df[column].quantile(0.75))
    IQR = Q3 - Q1

    maximum = Q3 + (1.5*IQR)

    maximum = df[column] > maximum

    return maximum

def Minimum(df, column):
    Q1 = (df[column].quantile(0.25))
    Q3 = (df[column].quantile(0.75))
    IQR = Q3 - Q1

    minimum = Q1 - (1.5*IQR)

    minimum = df[column] < minimum

    return minimum
```

```python
lower_PM25 = Minimum(changping_data_df, 'PM25')
more_PM25 = Maximum(changping_data_df, 'PM25')

changping_data_df = changping_data_df.assign(
    PM25 = lambda x: x['PM25'].mask(
        (x['PM25'] < lower_PM25) | (x['PM25'] > more_PM25),
        x['PM25'].rolling(window=10, min_periods=1).mean()
    )
)
```

```python
lower_PM10 = Minimum(changping_data_df, 'PM10')
more_PM10 = Maximum(changping_data_df, 'PM10')

changping_data_df = changping_data_df.assign(
    PM10 = lambda x: x['PM10'].mask(
        (x['PM10'] < lower_PM10) | (x['PM10'] > more_PM10),
        x['PM10'].rolling(window=10, min_periods=1).mean()))
```

### Pertanyaan 1: Bagaimana gambaran kualitas udara bulanan dan tahunan di kota tersebut?

```python
def create_monthly_date_df(df):
    monthly_date_df = df.resample(rule='M', on='datetime').agg({
        'No': 'nunique',
        'PM25': 'mean',
        'PM10': 'mean',
        'SO2': 'mean',
        'NO2': 'mean',
        'CO': 'mean',
        'O3': 'mean',
        'TEMP': 'mean',
        'PRES': 'mean',
        'DEWP': 'mean',
        'RAIN': 'mean',
        'WSPM': 'mean'
    })
    monthly_date_df.index = monthly_date_df.index.strftime('%B %Y')
    monthly_date_df = monthly_date_df.reset_index()
    monthly_date_df.rename(columns={
        'datetime': 'Bulan',
        'No': 'Jumlah'
    }, inplace=True)
    return monthly_date_df
```

```python
def create_yearly_date_df(df):
    yearly_date_df = df.resample(rule='Y', on='datetime').agg({
        'No': 'nunique',
        'PM25': 'mean',
        'PM10': 'mean',
        'SO2': 'mean',
        'NO2': 'mean',
        'CO': 'mean',
        'O3': 'mean',
        'TEMP': 'mean',
        'PRES': 'mean',
        'DEWP': 'mean',
        'RAIN': 'mean',
        'WSPM': 'mean'
    })
    yearly_date_df.index = yearly_date_df.index.strftime('%Y')
    yearly_date_df = yearly_date_df.reset_index()
    yearly_date_df.rename(columns={
        'datetime': 'Tahun',
        'No': 'Jumlah'
    }, inplace=True)
    return yearly_date_df
```

```python
monthly_changping_df = create_monthly_date_df(changping_data_df)
monthly_changping_df.head()
```

```python
yearly_changping_df = create_yearly_date_df(changping_data_df)
yearly_changping_df.head()
```

```python
sns.set(style='darkgrid')

plt.figure(figsize=(12,6))

sns.lineplot(data=monthly_changping_df, x='Bulan', y='PM25', label='PM2.5', color='blue')
sns.lineplot(data=monthly_changping_df, x='Bulan', y='PM10', label='PM10', color='green')
sns.lineplot(data=monthly_changping_df, x='Bulan', y='SO2', label='SO2', color='red')
sns.lineplot(data=monthly_changping_df, x='Bulan', y='NO2', label='NO2', color='purple')
sns.lineplot(data=monthly_changping_df, x='Bulan', y='O3', label='O3', color='orange')
sns.lineplot(data=monthly_changping_df, x='Bulan', y='TEMP', label='TEMP', color='brown')

plt.xlabel('Bulan')
plt.ylabel('Konsentrasi (g/m)')
plt.title('Rata-rata Pencemar udara setiap bulan')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def display_air_quality_graph(df, selected_city, selected_date, selected_hour):
    date_filtered = df[
        (df['datetime'].dt.date == selected_date) & (df['hour'] == selected_hour)
    ]

    city_parameters = {
        'Changping': ['PM25_changping', 'PM10_changping', 'SO2_changping', 'NO2_changping', 'CO_changping', 'O3_changping',
                      'TEMP_changping', 'PRES_changping', 'DEWP_changping', 'RAIN_changping', 'wd_changping', 'WSPM_changping'],
        'Dongsi': ['PM25_dongsi', 'PM10_dongsi', 'SO2_dongsi', 'NO2_dongsi', 'CO_dongsi', 'O3_dongsi',
                   'TEMP_dongsi', 'PRES_dongsi', 'DEWP_dongsi', 'RAIN_dongsi', 'wd_dongsi', 'WSPM_dongsi'],
        'Guanyuan': ['PM25_guanyuan', 'PM10_guanyuan', 'SO2_guanyuan', 'NO2_guanyuan', 'CO_guanyuan', 'O3_guanyuan',
                     'TEMP_guanyuan', 'PRES_guanyuan', 'DEWP_guanyuan', 'RAIN_guanyuan', 'wd_guanyuan', 'WSPM_guanyuan'],
        'Gucheng': ['PM25_gucheng', 'PM10_gucheng', 'SO2_gucheng', 'NO2_gucheng', 'CO_gucheng', 'O3_gucheng',
                    'TEMP_gucheng', 'PRES_gucheng', 'DEWP_gucheng', 'RAIN_gucheng', 'wd_gucheng', 'WSPM_gucheng'],
        'Average': ['PM25', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'wd', 'WSPM']
    }

    if not date_filtered.empty:
        parameters = city_parameters[selected_city]

        parameters_to_barplot = [param for param in parameters if 'wd' not in param]

        date = date_filtered[parameters_to_barplot].melt(var_name='Parameter', value_name='Value')

        sns.set(style='darkgrid')
        fig, ax = plt.subplots(figsize=(10, 6))
        sns.barplot(
            x='Value',
            y='Parameter',
            data=date,
            ax=ax,
            palette='viridis',
            hue='Parameter',
            legend=False
        )
        for container in ax.containers:
            ax.bar_label(container, fmt='%.2f')
        ax.set_title(f'Air Quality of {selected_city} City ({selected_date}-{selected_hour}:00)', fontsize=20)
        ax.set_xlabel('Value')
        ax.set_ylabel('Parameter')

        plt.show()
    else:
        print('Date not available')
```

```python

display_air_quality_graph(all_df, 'Dongsi', (datetime(2014, 8, 17).date()), 10)
```
